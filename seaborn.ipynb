```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
```

```python
plt.plot(x,y, marker='o', linestyle='-', color='b', label='Trend')
plt.xlabel('X_DATA')
plt.ylabel('Y_DATA')
plt.title('Line Plot Graph')
plt.legend()
plt.show()
```

# Bar Graph

```python
cat = ['A','B','C','D']
values = [30,50,20,80]
```

```python
plt.bar(cat, values, color=['blue', 'green', 'red', 'pink'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Categories and Values')
plt.show()
```

```python
sns.barplot(x=cat, y=values)
plt.show()
```

# Histogram

# BOX Plot

```python
df = pd.read_csv('titanic (1).csv')
```

```python
df.head()
```

```python
sns.boxplot(x = df['Fare'])
plt.title('BOX Plot On Fare')
plt.show()
```

# CORR

```python
cr_data= df[['Survived', 'Pclass', 'Age', 'Fare']]
```

```python
corelation = cr_data.corr()
sns.heatmap(corelation, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```
